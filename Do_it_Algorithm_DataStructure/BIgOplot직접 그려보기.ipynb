```python
from matplotlib import pyplot as plt
import numpy as np



def Big_O_plot(num):
    x = np.arange(1,num)
    one = [1 for _ in range(1, num)]
    log_n = np.log(x)
    n = x
    nlogn = (np.log(x))*10
    n2 = n ** 2
    n3 = n ** 3 
    two_power = [2 ** i for i in range(1, num)]

    plt.plot(x,one, label = '1')
    plt.plot(x,log_n, label = "log_n")
    plt.plot(x,n, label = "n")
    plt.plot(x,n2, label = 'n^2')
    plt.plot(x,n3, label = 'n^3')
#     plt.plot(x,two_power)
    plt.legend(loc = 'upper left')
    plt.show()
```

```python
Big_O_plot(5)
```

```python
Big_O_plot(10)
```
