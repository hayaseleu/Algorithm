````python
##1000 이하의 소수 나열 

counter = 0  # 나눗셈 횟수 카운ㅌ

for n in range(2,1001):
    for i in range(2, n):
        counter += 1
        if n % i == 0:
            break
    else: 
        print(n)
print(f'나눗셈 실행 횟수:{counter}')
````

````python
counter = 0 # 나눗셈 횟수 
ptr = 0     # 이미 찾은 소수 개수
prime = [None] * 500 # 소수 저장할 배열 

prime[ptr] = 2
ptr += 1 

for n in range(3,1001,2):
    for i in range(1,ptr):
        counter += 1
        if n % prime[i] == 0:
            break
    else:
        prime[ptr] = n
        ptr += 1

for i in range(ptr):
    print(prime[i]) 

print(f'나눗셈을 실행한 횟수: {counter}')
````

````python
# 알고리즘 개선 (대칭구조 활용 )

counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1 

prime[ptr] = 3
ptr += 1 

for n in range(5, 1001, 2):
    i = 1
    while prime[i] * prime[i] <= n:
        counter += 2
        if n % prime[i] == 0:
            break
        i += 1
    else:
        prime[ptr] = n
        ptr += 1
        counter += 1 

for i in range(ptr):
    print(prime[i])
print(f'(곱셈과 나눗셈을 실행한 횟수 : {counter}')
````
