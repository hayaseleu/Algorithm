정수 N이 입력되면서 00시 00분 00초 부터 N시 59분 59초까지의 모든 시각중에서 3이 하나라도 포함되는 모든 경우의 수를 구하는 프로그램을 작성하시오

```python
N = int(input())
```

## 첫번째 풀이

```python
def solution(N):
    tot = N* 3600 + 59 * 60 + 59
    count = 0
    while tot > 0:
        a = (tot // 3600)
        b = (tot - a * 3600) // 60
        c = (tot -a*3600- b*60) 
        printed = f'{a:#02}시 {b:#02}분 {c:#02}초'
        if '3' in str(a):
            count += 1
            tot -=1
            print(printed)
            continue
        if '3' in str(b):
            count += 1
            tot -= 1
            print(printed)

            continue
        if '3' in str(c):
            count += 1
            tot -= 1
            print(printed)
            continue
        tot -= 1
    return count
    
        
```

```python
solution(5)
```

## 두번째 풀이

```python
def solution_1(N):
    count = 0
    for i in range(N+1):
        for j in range(60):
            for k in range(60):
                if '3' in str(i) + str(j) + str(k):
                    count += 1
    return(count)
```
